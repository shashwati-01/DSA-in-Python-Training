LeetCode question    ## Edit Distance code also known as Levenshtein Distance

```python
s = "ycce"  ## Edit Distance code also known as Levenshtein Distance
t = "ycshashwatice"
output = 0

if len(s) > len(t):
    output = len(s) - len(t)

elif len(t) > len(s):
    output = len(t) - len(s)

elif len(s) == len(t):
    # For loop to check character by character when lengths are equal
    for i in range(len(s)):
        if s[i] != t[i]:
            output += 1      # This is the count++ logic for mismatches

print("Output:", output)
```
