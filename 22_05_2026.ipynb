# Hash Table

```python
# Python Implementation

class HashTable:
    def __init__(self, size):
        self.size = size

        # self.table = [[] for i in range(size)]

        self.table = []

        for i in range(size):
            self.table.append([])

    def hash_function(self, key):
        return key % self.size

    def insert(self, key=35):
        index = self.hash_function(key)
        self.table[index].append(key)

    def display(self):
        print(self.table)


h = HashTable(10)

h.insert(15)
h.insert(25)
h.insert(35)

h.display()
```

```python
# Linear Probing

class LinearProbing:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return key % self.size

    def insert(self, key):
        index = self.hash_function(key)

        while self.table[index] is not None:
            index = (index + 1) % self.size

        self.table[index] = key

    def display(self):
        print(self.table)


h = LinearProbing(10)

h.insert(15)
h.insert(25)
h.insert(35)

h.display()
```

```python
# Build Hash Table Manually

class PythonTable:

    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return key % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        self.table[index].append((key, value))

    def search(self, key):
        index = self.hash_function(key)

        for k, v in self.table[index]:
            if k == key:
                return v
        return "Not Found "

    def delete(self, key):
        index = self.hash_function(key)

        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                del self.table[index][i]
                return

    def display(self):
      print(self.table)

h = PythonTable(10)
h.insert(1, "pooja")
h.insert(11, " Shashwati")

print(h.search(11))

h.display()
h.delete(11)
h.display()
```

```python
class Solution:
    def twoSum(self, nums: list[int], target: int):
        hash_map = {}

        for i in range(len(nums)):
            current_num = nums[i]
            complement = target - current_num

            if complement in hash_map:
                return [hash_map[complement], i]

            hash_map[current_num] = i
```

```python
# search() function
# we can use search function to check the given pattern in
# beginning of target string.
# if the match is available then we will get match object,
# otherwise we will get None.

import re

s = input("Enter any string: ")
m = re.search("python", s)

if m != None:
    print("The string is available at begin")
    print("start index:", m.start(), "and last:", m.end())
else:
    print("String is not available at begin")
```

```python
# We can use fullmatch() function to match a pattern to
# all of target string. The complete string should be
# matched according to given pattern.

import re

str = input("Enter any string : ")

m = re.fullmatch(str, "ababaabc")

if m != None:
    print("matching is available")
else:
    print("matching is not available")
```
